```python
from scapy.all import *
import random
import time
```

```python
# send reward package
def send_reward_packet(packet_num=500, Sleep=0.5, Dst='127.0.0.1', Dport=80, Flags='S'):
    total_packet_num = packet_num
    while total_packet_num:
        random_ip = '%i.%i.%i.%i' % (random.randint(1, 254),random.randint(1, 254),random.randint(1, 254),random.randint(1, 254))
        ip = IP(src=random_ip, dst=Dst, ttl=80)
        packet = TCP(dport=Dport, flags=Flags)
        payload = "1"
        send(ip/packet/payload)
        total_packet_num -= 1
        time.sleep(Sleep)
```

```python
for i in range(2):
    send_reward_packet(packet_num=200, Sleep=random.randint(1,3)/10)
```
